```python
# plot to show  each team hitting   with total runs
df_bat.groupby('Team')['Runs'].sum().plot(kind='bar',title='Teams VS Total runs')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.show()
```

```python
# pie chart : percentage distribution
df_bat.groupby('Team')['Runs'].sum().plot(kind='pie',autopct='%1.0f%%',title='Total Teams VS Total runs')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
# plot to show  each team  taking  their total   wickets  in whole season of ipl
df_bowl.groupby('Team')['WKT'].sum().plot(kind='bar',title='Teams VS Total wickets')
plt.xlabel('Teams')
plt.ylabel('Total Wickets')
plt.show()
```

```python
# plot to show  each team taking total wickets
df_bowl.groupby('Team')['WKT'].sum().plot(kind='pie',autopct='%1.0f%%',title='Total Teams VS Total wickets')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
# Top 3 run hitters in  whole ipl season [for orange cap]
o_c=df_bat.nlargest(3,'Runs').plot(kind='bar',x='Player Name' , y='Runs',figsize=(10, 8),title='Top  run Scorers for orange cap',color='orange')
# Adding  run values on top of bars
# --- This is the key change: Using variablename.bar_label() ---
# variablename.containers[0] gets the BarContainer object for the first (and only) set of bars
# fmt for integer, padding to lift labels
o_c.bar_label(o_c.containers[0], fmt='%.0f', padding=5)
plt.xlabel('Player name')
plt.ylabel('Runs')
plt.show()
```

```python
# Top 3 run hitters in  whole ipl season [for purple cap]
p_c=df_bowl.nlargest(3,'WKT').plot(kind='bar',x='Player Name', y='WKT',figsize=(10, 8),title='Top  Wicket Taker for purple cap',color='purple')
# Adding  no of wicket taken  by player  on top of bars
# --- This is the key change: Using variablename.bar_label() ---
# variablename.containers[0] gets the BarContainer object for the first (and only) set of bars
# fmt for integer, padding to lift labels
p_c.bar_label(p_c.containers[0], fmt='%.0f', padding=5)
plt.xlabel('Player name')
plt.ylabel('WKT')
plt.show()
```

```python
# Top 3 run hitters with thier team name  in  whole ipl season [for orange cap]
# Prepare the data (using two variables)
o_c=df_bat.nlargest(3,'Runs')
o_c['Player_Team'] = o_c['Player Name'] + ' (' + o_c['Team'] + ')'
#using 2nd variable
o_cc=o_c.plot(kind='bar',x='Player_Team' , y='Runs',figsize=(10, 8),title='Top 3 run Scorers',color='orange')
o_cc.bar_label(o_cc.containers[0],fmt='%.0f', padding=5)
plt.xlabel('Player with team')
plt.ylabel('Runs')
plt.show()
```

```python
# Top 3 wicket taker with thier team name  in  whole ipl season [for purple cap]
# Prepare the data (using two variables)
p_c=df_bowl.nlargest(3,'WKT')
#for diplaying  player name and team name
p_c['Player_Team'] = p_c['Player Name'] + ' (' + p_c['Team'] + ')'
#using 2nd variable
p_cc=p_c.plot(kind='bar',x='Player_Team' , y='WKT',figsize=(10, 8),title='Top 3 wicket taker',color='purple')
p_cc.bar_label(p_cc.containers[0],fmt='%.0f', padding=5)
plt.xlabel('Player with team')
plt.ylabel('Wickets')
plt.show()
```

```python
#to show the  anaylsis between no of matches played vs run scored  by overall team in scatter plot
sb.scatterplot(x='Matches',y='Runs',data=df_bat)
plt.show()
```

```python
#finding if there is a correlation between the number of matches played and the total runs scored.
bat_cor = df_bat.select_dtypes(include=np.number)
bat_cor.corr()
```

```python
#Now creating the heat map  of  the correlation to find relation between the columns
sb.heatmap(bat_cor.corr(),
           annot=True,       # Annotate the heatmap with the correlation values
           cmap='coolwarm',  # Choose a color map (coolwarm is great for correlations)
           fmt=".2f",        # Format the annotations to 2 decimal places
           linewidths=.5)    # Add lines to separate the cells
plt.title('Correlation map of IPL Batting Statistics', fontsize=16)
plt.show()
```

```python
#comparing  average and strike rate .
sb.scatterplot(x='AVG',y='SR',data=df_bat)
plt.show()
```

```python
#distribution of average and strike rate by boxplot.
sb.boxplot(x='AVG',y='SR',data=df_bat)
plt.show()
```

```python
#Visualization  for the most boundary hitting   team in bar chart
ax=df_bat.groupby('Team')['Boundary'].sum().plot(kind='bar',title='Teams VS Total Boundary',x='Team' , y='Boundary',figsize=(10, 8),color='green')
ax.bar_label(ax.containers[0],fmt='%.0f', padding=5)
plt.xlabel('Team')
plt.ylabel('Total Boundary')
plt.show()
#Acoording to insights  punjab kings relies on more boundaries
```

```python
#top 5 boundary hitters showing in bar chart now
most_boundaries=df_bat.nlargest(5,'Boundary').plot(kind='bar',x='Player Name' , y='Boundary',figsize=(10, 8),title='Top boundary hitters',color='green')
most_boundaries.bar_label(most_boundaries.containers[0],fmt='%.0f', padding=5)
plt.xlabel('Team')
plt.ylabel('Boundaries')
plt.show()
```

```python
#compare players and see if they prefer hitting along the ground or over the top. by scatter plot.
sb.scatterplot(x='4s',y='6s',data=df_bat)

# Add player names as labels
for index, row in df_bat.iterrows():
    plt.annotate(row['Player Name'], (row['4s'], row['6s']), textcoords="offset points", xytext=(0,10), ha='center', fontsize=6)

plt.title('4s vs 6s with Player Names')
plt.xlabel('Number of 4s')
plt.ylabel('Number of 6s')
plt.show()
```

```python
#Visualization  for the most four  wicket hall by   any  team in bar chart
fw=df_bowl.groupby('Team')['4W'].sum().plot(kind='bar',title='Teams VS Total 4 wicket haul',x='Team' , y='4W',figsize=(10, 8),color='green')
fw.bar_label(fw.containers[0],fmt='%.0f', padding=5)
plt.xlabel('Team')
plt.ylabel('Total 4 wicket hall taken')
plt.show()
#According to insights  mumbai india players  have taken  more four wicket haul
```

```python
#visualizing  who have hit more no of  centuries and half centuries.
sb.scatterplot(x='100s',y='50s',data=df_bat)
#Adding player name
for index, row in df_bat.iterrows():
    plt.annotate(row['Player Name'], (row['100s'], row['50s']), textcoords="offset points", xytext=(0,10), ha='center', fontsize=10)
plt.show()
```

```python
#visualization to see the highest score hit by players  by histogram plot
df_bat['Runs'].hist()
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.title('Distribution of Runs')
plt.show()
```

```python
#visualization for highest score not out by any player in whole season
# Convert 'HS' to numeric by removing the '*' and coercing errors
df_bat['HS_numeric'] = df_bat['HS'].str.replace('*', '', regex=False).astype(int, errors='ignore')
# Filter out rows where conversion failed (optional, depending on how you want to handle non-numeric 'HS')
hs = df_bat[df_bat['HS_numeric'].notna()]
# Now use nlargest on the new numeric column
hs=df_bat.nlargest(3,'HS_numeric').plot(kind='bar',x='Player Name' , y='HS_numeric',figsize=(10, 8),title='Top 3 Highest Scores',color='pink')
hs.bar_label(hs.containers[0],fmt='%.0f', padding=5)
plt.xlabel('Players')
plt.ylabel('Highest Score')
plt.show()
```

```python
# Good  economy comparision on bowlers
te=df_bowl.nsmallest(8,'ECO').plot(kind='bar',x='Player Name' , y='ECO',figsize=(10, 8),title='Good Economy bowlers',color='yellow')
te.bar_label(te.containers[0],padding=5)
plt.xlabel('Player')
plt.ylabel('Economy rate')
plt.show()

```

```python
#visualization for the  overs bowled by bowler and run given
plt.figure(figsize=(10, 6))
sb.scatterplot(x='OVR', y='RUNS', data=df_bowl)

# Add player names as labels
for index, row in df_bowl.iterrows():
    plt.annotate(row['Player Name'], (row['OVR'], row['RUNS']), textcoords="offset points", xytext=(0,10), ha='center', fontsize=6)

plt.title('Overs Bowled vs Runs Given with Player Names')
plt.xlabel('Overs Bowled')
plt.ylabel('Runs Given')
plt.show()
```
